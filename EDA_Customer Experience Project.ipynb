# 3.0 Data Exploration and Visualization

##### Amount paid for goods

```python
#Distribution of amount paid by customer

plt.figure(figsize=(10,4))
plt.subplot(2, 2, 1)
sns.histplot(data=sale, x='Amount',  palette='viridis')
plt.title('Distribution of amount paid by customer');

plt.subplot(2, 2, 2)
sns.boxplot(data=sale, x='Amount',  palette='viridis')
plt.title('Amount paid by customer');
```

##### Percentage distibution of products sold

```python
count_data = sale['Product'].value_counts().sort_values(ascending=False)
products = count_data.index

plt.figure(figsize=(6,4))
ax=sns.countplot(data = sale, x = 'Product', order=products, palette='viridis')
plt.title("Percentage distibution of products sold")

total = len(sale['Product'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.show()
```

- The order of product sale is laptop, tablet, headphones, smartphone and smartwatch

##### Percent sale per product type

```python
total_sales_per_product=sale[['Product', 'Amount']]
total_sales_per_product = sale.groupby('Product')['Amount'].sum()

plt.figure(figsize=(6,3))
sns.set(style="darkgrid")
ax = sns.barplot(x=total_sales_per_product.index, y=total_sales_per_product.values, palette='viridis')

total = sum(sale['Amount'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.xlabel('Product Type')
plt.ylabel('Total Sales')
plt.title('Percent Sales per Product Type')
plt.xticks(rotation=45, ha='right')
plt.show()
```

- Percentage income by product type shows laptop, tablet and headphones as top earners

```python
city_data = new_sale['City'].value_counts().sort_values(ascending=False).head(10)
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=city_data.index, y=city_data.values, palette="viridis")

plt.xlabel('City')
plt.ylabel('Count')
plt.title('Top 10 Cities by Count')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.xticks(rotation=45, ha='right');
```

##### Total sales per cities and countries

```python
# Calculate total sales per City and Country
total_sales_per_city = new_sale.groupby('City')['Amount'].sum().sort_values(ascending=False).head(10).reset_index()
total_sales_per_country = new_sale.groupby('Country')['Amount'].sum().sort_values(ascending=False).head(10).reset_index()

# Create subplots
plt.figure(figsize=(15, 5))

# Subplot for total sales per city
plt.subplot(1, 2, 1)
sns.barplot(data=total_sales_per_city, x='City', y='Amount',palette="viridis")
plt.xlabel('Cities')
plt.ylabel('Total Sales')
plt.title('Total Sales per Cities')
plt.xticks(rotation=45, ha='right')

# Subplot for total sales per country
plt.subplot(1, 2, 2)
sns.barplot(data=total_sales_per_country, x='Country', y='Amount',palette="viridis")
plt.xlabel('Countries')
plt.ylabel('Total Sales')
plt.title('Total Sales per Countries')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

##### Time Data

```python
#Extract the year, month and quarter
new_sale['Year'] = new_sale['TransactionDate'].dt.year
new_sale['Month'] = new_sale['TransactionDate'].dt.month
new_sale['Month_name'] = new_sale['TransactionDate'].dt.month_name()
new_sale['Time'] = new_sale['TransactionDate'].dt.time
```

```python
def get_time_of_day(x):
    if x >= pd.to_datetime('00:00:00').time() and x < pd.to_datetime('07:00:00').time():
        return 'Night'
    elif x >= pd.to_datetime('07:00:00').time() and x < pd.to_datetime('12:00:00').time():
        return 'Morning'
    elif x >= pd.to_datetime('12:00:00').time() and x < pd.to_datetime('18:00:00').time():
        return 'Afternoon'
    elif x >= pd.to_datetime('18:00:00').time() and x <= pd.to_datetime('23:59:59').time():
        return 'Evening'
    else:
        return 'Dawn'
    

new_sale['day_period'] = new_sale['Time'].apply(get_time_of_day)
```

##### Sales over the month and years

```python
# Group by year and month and calculate total sales
sales_over_time = new_sale.groupby(['Year', 'Month'])['Amount'].sum()

# Convert the result to a DataFrame
sales_over_time = sales_over_time.reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(sales_over_time['Year'].astype(str) + '-' + sales_over_time['Month'].astype(str), sales_over_time['Amount'], marker='o',
        )
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.title('Sales Over Months and Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

- Highest sale was observed in the 3rd month of 2023 and the least sale in the january of 2023

##### Sales by time of the day

```python
total_sales_per_day_time = round(new_sale.groupby('day_period')['Amount'].sum().sort_values(ascending=False).reset_index(),0)
total_sales_per_day_time  

sns.set(style="darkgrid")
plt.figure(figsize=(8, 4))
ax = sns.barplot( data = total_sales_per_day_time , x='day_period', y='Amount', palette = 'viridis')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')


plt.xlabel('Day Period')
plt.ylabel('Total Sales')
plt.title('Total Sales per Day Period')
plt.xticks(rotation=45, ha='right')
plt.show()
```

###### Merge of DataFrame with the country-continent mapping dataset

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
continent_df = pd.merge(new_sale, world[['name', 'continent']], left_on='Country', right_on='name', how='left')
#filling up missing countries as small islands
continent_df['continent']=continent_df['continent'].fillna('small Islands')
# Plot the results
plt.figure(figsize=(8, 4))
ax=sns.barplot(x='continent', y='Amount', 
               data=continent_df.groupby('continent')['Amount'].sum().reset_index(), palette = 'viridis')

plt.title('Amount by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Amount')
plt.xticks(rotation=45, ha='right')
plt.show()
```
